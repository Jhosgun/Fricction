```python
v_p = datosp.loc[:,'v_p']
v_p = v_p.values
a_p = datosp.loc[:,'a_p']
a_p = a_p.values
```

```python
k = []
g = 9.8
m = 0.0017
for i in range(1,len(a_p)-1) :
  K=m*(-g-a_p[i])/v_p[i]
  if K == math.inf:
    k.append(0)
  else:
    k.append(K)
```

```python
print(mean(k))
```

```python
v_s = np.empty(len(datosp.y_p))
y_s = np.empty(len(datosp.y_p))
a = np.empty(len(datosp.y_p))
g = -9.8
m = 0.017
const = 0.14
for i in range(0,len(datosp.y_p)-1):
  a[i+1] = g-(const/m)*v_s[i]
  v_s[i+1]=v_s[i]+a[i]*(datosp.t_p[i+1]-datosp.t_p[i])
  y_s[i+1]=y_s[i]+v_s[i]*(datosp.t_p[i+1]-datosp.t_p[i])+a[i]*((datosp.t_p[i+1]-datosp.t_p[i])**2/2)
#np.set_printoptions(formatter={'float_kind':'{:f}'.format})
print(v_s)
```
